```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn.decomposition import PCA
import warnings
from scipy.stats import gaussian_kde


# Importamos la base de datos
vinos=pd.read_excel(r'/Users/varantxa/Desktop/TFG/Wine_data.xlsx', sheet_name='White Wine')

# Hacemos un df de la base de datos
df=pd.DataFrame(vinos)

df=df.rename(columns = {'fixed acidity' : 'fixed_acidity', 'volatile acidity': 'volatile_acidity', 'citric acid' : 'citric_acid', 'residual sugar' : 'residual_sugar', 'free sulfur dioxide': 'free_sulfur_dioxide', 'total sulfur dioxide' : 'total_sulfur_dioxide'})

# Especificar el número de columnas en el grid
num_cols = 3

# Calcular el número de filas en función del número de variables y el número de columnas
num_rows = (len(df.columns) + num_cols - 1) // num_cols

# Crear el grid de subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 10))

# Iterar sobre las columnas y crear un box plot en cada subplot del grid
for i, column in enumerate(df.columns):
    row = i // num_cols
    col = i % num_cols
    ax = axes[row, col]
    sns.boxplot(data=df[column], ax=ax)
    ax.set_title(column)

# Ajustar el espaciado entre los subplots
plt.tight_layout()

# Mostrar el grid de box plots
plt.show()
    
#density plot

df.fixed_acidity.plot.density(color='green')
plt.title('Density plot for fixed acidity')
plt.show()
df.volatile_acidity.plot.density(color='blue')
plt.title('Density plot for volatile acidity')
plt.show()
df.citric_acid.plot.density(color='orange')
plt.title('Density plot for citric acid')
plt.show()
df.residual_sugar.plot.density(color='purple')
plt.title('Density plot for residual sugar')
plt.show()
df.free_sulfur_dioxide.plot.density(color='black')
plt.title('Density plot for free sulfur dioxide')
plt.show()
df.total_sulfur_dioxide.plot.density(color='yellow')
plt.title('Density plot for total sulfur dioxide')
plt.show()
df.quality.plot.density(color='red')
plt.title('Density plot for quality')
plt.show()
df.pH.plot.density(color='grey')
plt.title('Density plot for pH')
plt.show()
df.chlorides.plot.density(color='beige')
plt.title('Density plot for chlorides')
plt.show()
df.alcohol.plot.density(color='pink')
plt.title('Density plot for alcohol')
plt.show()
df.density.plot.density(color='lightblue')
plt.title('Density plot for density')
plt.show()
df.sulphates.plot.density(color='lightgreen')
plt.title('Density plot for sulphates')
plt.show()

# Hacemos un pairplot para ver cómo se relaciona cada variable entre ellas 
scaler = StandardScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])
sns.pairplot(df)
plt.show()

# Calcula la matriz de correlación
correlation_matrix = df.corr()

# Imprime la matriz de correlación
print(correlation_matrix)


```
